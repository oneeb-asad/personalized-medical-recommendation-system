# Step 1: Data Preprocessing and Model Training
This notebook is designed to walk through the first phase of our project — loading the dataset, cleaning it, encoding labels, and training a baseline machine learning model using Random Forest.

**Dataset Used**: `Training.csv` (from Kaggle dataset folder).

**Goal**: Predict diseases based on symptoms using machine learning.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import pickle
```

## Load the Training Dataset
We begin by loading `Training.csv`, which includes binary symptom data and a `prognosis` column representing the disease.

```python
# Load dataset
df = pd.read_csv('../data/Training.csv')
print("Dataset shape:", df.shape)
df.head()
```

## Check for Null or Missing Values
A clean dataset is essential for training a model.

```python
# Check for missing values
df.isnull().sum().sum()
```

## Encode Target Variable (Prognosis)
We convert disease names into numeric labels using `LabelEncoder`.

```python
# Separate features and target
X = df.drop('prognosis', axis=1)
y = df['prognosis']

# Encode the target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
#created the  directory
import os
os.makedirs('../models', exist_ok=True)
# Save the label encoder for Flask app usage later
with open('../models/disease_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)
import os
if os.path.exists('../models/disease_encoder.pkl'):
    print("Label encoder saved successfully at ../models/disease_encoder.pkl")
```

## Split the Dataset
We split the data into training and testing subsets (80/20).

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

## Train the Random Forest Model
Random Forest is a robust and accurate classifier that works well for this type of data.

```python
# Train the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

## Evaluate the Model
We'll evaluate performance using accuracy and classification metrics.

```python
# @title Default title text
# Predictions and evaluation
y_pred = rf_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Save the Trained Model
We’ll use this model later in the Flask application to predict disease from user symptoms.

```python
# Save the trained model
with open('../models/random_forest.pkl', 'wb') as f:
    pickle.dump(rf_model, f)
# download the file ../models/rf_model.pkl

import os
if os.path.exists('../models/random_forest.pkl'):
    print("Random Forest model saved successfully at ../models/random_forest.pkl")
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

# Train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Evaluate the Decision Tree model
y_pred_dt = dt_model.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))

# Save the trained Decision Tree model
with open('../models/decision_tree.pkl', 'wb') as f:
    pickle.dump(dt_model, f)

import os
if os.path.exists('../models/decision_tree.pkl'):
    print("Label encoder saved successfully at ../models/decision_tree.pkl")
```

# MLPClassifier

```python
from sklearn.neural_network import MLPClassifier

# Train the MLP model
mlp_model = MLPClassifier(random_state=42, max_iter=500)
mlp_model.fit(X_train, y_train)

# Evaluate the MLP model
y_pred_mlp = mlp_model.predict(X_test)
print("MLP Accuracy:", accuracy_score(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))

# Save the trained MLP model
with open('../models/neural_network.pkl', 'wb') as f:
    pickle.dump(mlp_model, f)

import os
if os.path.exists('../models/mlp_model.pkl'):
    print("Label encoder saved successfully at ../models/neural_network.pkl")
```
